```python
df = pd.read_excel(path + "stoxx600.xlsx")

df.rename(columns={df.columns[0]: "Date"}, inplace=True)
df["Date"] = pd.to_datetime(df["Date"], errors="coerce")
```

```python
df = df.loc[:, ~df.columns.str.startswith("#ERROR")]
```

```python
df_long = df.melt(
    id_vars="Date",
    var_name="col",
    value_name="value"
)
```

```python
df_long[["Stock", "Datatype"]] = (
    df_long["col"]
    .str.split(" - ", n=1, expand=True)
)
```

```python
df_panel = (
    df_long
    .pivot_table(
        index=["Date", "Stock"],
        columns="Datatype",
        values="value"
    )
    .reset_index()
)
```

```python
df_panel = df_panel.sort_values(["Stock", "Date"])

print(df_panel.head())
print(df_panel.tail())
print(df_panel.info())
```

```python
df_panel["Stock"].nunique()
```

```python
df_panel["RET"] = (
    df_panel.groupby("Stock")["TOT RETURN IND"]
    .pct_change()
)
```
